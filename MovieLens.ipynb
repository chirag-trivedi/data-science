```python
rating_list = ['Action', 'Adventure',
       'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
movie[rating_list].sum().sort_values(ascending=False)
```

Here are the genres and the counts sorted in desc order

**Can we find the movies that have more than one genre?**

```python
movie['Number of Genres'] = movie.loc[:,'Action':'Western'].sum(axis=1)
movie.head()
```

```python
movie[movie['Number of Genres']>1][['movie title','Number of Genres']].sort_values(ascending=False,by='Number of Genres').head(5).style.hide()
```

We add a column that counts the no of genres

**Can we find the genres that are most liked by the users?**

```python
df_merge = movie.merge(rating,on='movie id',how="inner")
df_merge.head()
```

```python
rating_list
```

```python
for i in rating_list:
    print(i,':' , df_merge[df_merge[i]==1].rating.mean())
```

Can we find which movies are rated most of the times ?

```python
movie_count = df_merge.groupby(['movie title'])['rating'].count().reset_index()
movie_count.sort_values(ascending=False,by='rating').head(5).style.hide()
```

Star Wars,Contact and Fargo are the top 3 rated most

extract some relationship between demographic details of the users and ratings for the movies?

```python
df_user_rating = user.merge(rating,on='user id',how='inner')
df_user_rating.groupby(['gender'])['rating'].mean()
```

Split between male and female is pretty much the same for ratings

```python
df_user_rating.groupby(['occupation'])['rating'].mean()
```

Splits by jobs

```python

df_user_rating_sorted = df_user_rating.groupby(['movie id','zip code'])['rating'].mean().sort_values(ascending=False).to_frame()

df_user_rating_sorted.merge(movie,on='movie id',how='inner')[['movie title','rating']].head(5).style.hide()
```
